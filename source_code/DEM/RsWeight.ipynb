```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
pci_dist = 10
hole = 40
```

```python
path = "/root/opt/exp_result/p-{}mm/e-{}mm".format(pci_dist, hole)
df = pd.read_csv(path, usecols=["Time","Data"])
print(df)
```

```python
df["Time"] = pd.to_datetime(df["Time"], format = "%H:%M:%S")
df = df.groupby('Time', group_keys=False, as_index=False).apply(lambda x: x.loc[x.Data.idxmax()])
print(df)
```

```python
serial_num = pd.RangeIndex(start=0, stop=len(df.index), step=1)
df['time'] = serial_num
print(df)
```

```python
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot()
#ax.set_title("Time - Quantity",fontsize=24)
ax.set_xlabel("Time[s]",fontsize=24)
ax.set_ylabel("Quantity",fontsize=24)
ax.tick_params(axis='x',labelsize=16)
ax.tick_params(axis='y',labelsize=16)

x = df["time"]
y = df["Data"]
ax.plot(x,y, '1', color='deepskyblue', label='simulation')

ax.legend(loc='lower right', fontsize=24)

plt.show()
```
