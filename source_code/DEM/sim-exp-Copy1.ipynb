```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import glob
from sklearn.linear_model import LinearRegression
```

```python
pci_dist = 5
holesize = 40
cuttime = 15
```

```python
def readcsv(path):
    DF = pd.read_csv(path, usecols=["Time","Data"])
    DF = DF.dropna()
    DF = DF[:-1]
    DF = DF[(len(DF[DF.Data==0.0]) - 1):]
    DF["Time"] = pd.to_datetime(DF["Time"], format = "%H:%M:%S")
    DF = DF.groupby('Time', group_keys=False, as_index=False).apply(lambda x: x.loc[x.Data.idxmin()])
    DF["time"] = 0.0
    for i in range(len(DF)):
        DF.time[i:(i+1)]=i
    return DF
```

```python
name = [''] * 10
for i in range(0,10):
    x=str(i+1)
    path = "/root/opt/exp_result/p-{}mm/e-{}mm/{}mm-{}.csv".format(pci_dist, holesize, holesize, x.zfill(2))
    print(path)
    name[i]=readcsv(path)
    print(name[i])
```

sim読み込み

```python
path = "/root/opt/sim_result/p-{}mm/{}mm".format(pci_dist, holesize)
filename = "/output"
filetype = ".csv"

dt = 0.1
D = pci_dist / 1000
dns = 2500
volume = dns * np.pi * D*D*D / 6
under = 0.35

csv_files =glob.glob(path+ filename+ "*" + filetype)
#print(csv_files)
print("総ファイル数は" + str(len(csv_files)))
```

読み込みファイル設定

10mmはpos.y, pos.z, velocityのカラムの頭に空白入ってる

```python
time_amount = pd.DataFrame(columns=["time","quantity", "mass"])#時間と流出個数
for i in range(0,len(csv_files)-2): #wall and seitei
    openfile = path + filename + str(i).zfill(5) +filetype     
    df = pd.read_csv(openfile, usecols=["Pos.x","Pos.y","Pos.z","velocity"])
    df2 = df[df["Pos.y"]<under]#under以下の個数カウント
    add = pd.DataFrame([[i*dt, len(df2), len(df2)*volume]], columns = ["time","quantity","mass"])
    time_amount = time_amount.append(add)
    #print(openfile)
    
del(df)
del(df2)
del(add)
```

ファイル順番に読み込んで処理

```python
print(time_amount)
```

```python
time_amount = time_amount[time_amount.time<=cuttime]
```

plot

```python
import matplotlib.cm as cm
def Plot(data,n,Fig):
    N=str(n)
    data = data[data.time<=cuttime]
    X = data["time"]
    Y = data["Data"]
    Fig.plot(X,Y, 's', ms=8, color="green")
```

```python
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot()
#ax.set_title("Hole-{}mm".format(holesize),fontsize=24)
ax.set_xlabel("Time[s]",fontsize=32)
ax.set_ylabel("Weight[g]",fontsize=32)
ax.tick_params(axis='x',labelsize=24)
ax.tick_params(axis='y',labelsize=24)

ax.set_xlim(-1,17)
ax.set_ylim(-500,10500)

for i in range(0,10):
    Plot(name[i],i+1,ax)

sim_plot = time_amount[0::10]
xp = sim_plot["time"]
yp = sim_plot["mass"]
ax.plot(xp,yp*1000, 'o', ms=10, color='red', label='simulation')
ax.plot(100,15000, 's', ms=8, color="green",label="experiment")

ax.legend(loc='lower right', fontsize=32)

plt.show()
```
