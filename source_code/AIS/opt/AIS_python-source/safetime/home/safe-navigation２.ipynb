```python
import numpy as np
import pandas as pd
import glob
import pygrib
#%matplotlib inline
import matplotlib.pyplot as plt
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import cartopy.crs as ccrs
import datetime as dt
from IPython.display import clear_output
```

```python
path = "D:/AIS/AIS-Container_Data2.csv"
```

```python
frame = pd.read_csv(path,encoding='UTF-8') 
frame
```

```python
frame2 = frame[( (frame["longitude"]>150)&(frame["longitude"]<180) | (frame["longitude"]>-180)&(frame["longitude"]<-135) ) & (frame["latitude"]>0)&(frame["latitude"]<47)]
```

```python
fig = plt.figure(figsize=(160,90), dpi=12)

ax = fig.add_subplot(  1, 1, 1, projection=ccrs.PlateCarree(central_longitude=0))
ax.set_global()
ax.stock_img()
ax.coastlines()

X1 = frame2.longitude
Y1 = frame2.latitude
plt.scatter(X1,Y1, c='red',s=0.1)
```

```python
fig.savefig("D:/AIS/safe-plot/speed/container_speed.jpg")
```

```python
frame2.Fr.describe()
```

```python
frame3 = frame2[frame2.Fr >= 0.05]
frame3.describe()
```

froude length

```python
fig2 = plt.figure()

ax2 = fig2.add_subplot(1, 1, 1)
ax2.set_xlabel("Froude Number")
ax2.set_ylabel("Frequency")

frame4 = frame3[(frame3.length >= 300) & (frame3.length < 10000)].Fr
frame4 = frame4.round(5)
n, bins, patches = plt.hist(frame4,bins=100, range=[0.05,0.3])

mode_index = n.argmax()

ax2.text(0.21,23000, "Median : " + str(frame4.median()))
ax2.text(0.21,21000, "Mode : " + str(bins[mode_index].round(5)) + " - " + str(bins[mode_index+1]))

plt.title("Length    over 300m")
fig2.savefig("D:/AIS/safe-plot/speed/300/Froude2_over300.jpg")
```
